```python
#average of normalized loss column
avg_norm_loss = df['normalized-losses'].astype("float").mean(axis=0)
print(avg_norm_loss)
```

```python
#replaced NaNs with average 
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
df["normalized-losses"].head(10)
```

```python
#calcs average bore
avg_bore_df = df["bore"].astype('float').mean(axis=0)
print(avg_bore_df)
```

```python
#replaces Nans in horsepower and peak rpm columns with averages; really should do average by vehicle type
#note that axis = 0 denotes a change down the horsepower column in the y direction
avg_horsepower = df["horsepower"].astype('float').mean(axis=0)

df["horsepower"].replace(np.nan, avg_horsepower, inplace = True)


avg_peakrpm = df["peak-rpm"].astype('float').mean(axis=0)

df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace = True)
```

```python
#grouping columns
df_group_one = df3[['drive-wheels', 'body-style','price']]

df_group_one.tail(5)
```

```python
#groups vehicles by drivewheel and calculates average price per drive wheel type
df_group_one = df_group_one.groupby(['drive-wheels'], as_index=False).mean()
df_group_one
```

```python
#gives average price data based on grouping of hatchback and drivewheel
df_group_two = df[['drive-wheels', 'body-style','price']]
gptest = df_group_two.groupby(['drive-wheels', 'body-style'],as_index=False).mean()
gptest
```

```python
#pivot table displays the data better
gp_pivot = gptest.pivot(index='drive-wheels',columns='body-style')
gp_pivot
```

```python
#fills in NaN data points
gp_pivot = gp_pivot.fillna(0)
gp_pivot
```

```python
gp_pivot.dtypes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.pcolor(gp_pivot, cmap = 'RdBu')
plt.colorbar()
plt.show()
```

```python
#setting up the plot above so it looks better

fig, ax = plt.subplots()
im = ax.pcolor(gp_pivot, cmap='RdBu')

#label names
row_labels = gp_pivot.columns.levels[1]
col_labels = gp_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(gp_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(gp_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
